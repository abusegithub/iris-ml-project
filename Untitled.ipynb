```python
# Cell 2 - imports & output dirs
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold, learning_curve
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder, label_binarize
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             classification_report, confusion_matrix, roc_curve, auc)
import joblib

# Paths (update if your base path differs)
BASE_PATH = r"C:\ML PROJECTS\1. Iris Flower Classification\IRISS Dataset"
OUT_DIR = os.path.join(os.path.dirname(BASE_PATH), "output")
FIG_DIR = os.path.join(OUT_DIR, "figs")
MODEL_DIR = os.path.join(OUT_DIR, "models")

os.makedirs(FIG_DIR, exist_ok=True)
os.makedirs(MODEL_DIR, exist_ok=True)

sns.set(style="whitegrid")
print("Output folders:", OUT_DIR, FIG_DIR, MODEL_DIR)
```

```python
# Cell 3 - auto-detect dataset file (csv or xlsx)
possible = glob.glob(os.path.join(BASE_PATH, "iris.*")) + glob.glob(os.path.join(BASE_PATH, "Iris.*")) + glob.glob(os.path.join(BASE_PATH, "IRIS.*"))
possible = list(dict.fromkeys(possible))  # unique

if not possible:
    raise FileNotFoundError(f"No iris.* file found in {BASE_PATH}. Put iris.csv or iris.xlsx there.")

DATA_FILE = possible[0]
print("Using dataset file:", DATA_FILE)

# Load depending on extension
if DATA_FILE.lower().endswith(".csv"):
    df = pd.read_csv(DATA_FILE)
elif DATA_FILE.lower().endswith((".xls", ".xlsx")):
    df = pd.read_excel(DATA_FILE)
else:
    # fallback: try sklearn dataset
    iris = datasets.load_iris()
    df = pd.DataFrame(iris.data, columns=iris.feature_names)
    df['species'] = [iris.target_names[i] for i in iris.target]
    print("Loaded built-in sklearn iris as fallback.")

# Quick preview
print("Columns:", df.columns.tolist())
display(df.head())
```

```python
# Cell 4 - normalize column names and check content
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

# Common expected names mapping (in case file has different headers)
expected = ["sepal_length","sepal_width","petal_length","petal_width","species"]
# attempt to rename common variants
col_map = {}
for c in df.columns:
    lc = c.lower()
    if "sepal" in lc and "length" in lc: col_map[c] = "sepal_length"
    if "sepal" in lc and "width" in lc: col_map[c] = "sepal_width"
    if "petal" in lc and "length" in lc: col_map[c] = "petal_length"
    if "petal" in lc and "width" in lc: col_map[c] = "petal_width"
    if "species" in lc or "class" in lc or "target" in lc: col_map[c] = "species"
if col_map:
    df = df.rename(columns=col_map)

missing = [c for c in expected if c not in df.columns]
if missing:
    raise ValueError(f"Missing columns in dataset: {missing}. Found columns: {df.columns.tolist()}")

print("Final columns:", df.columns.tolist())
print("Sample values for species:", df['species'].unique())
display(df.head())
```

```python
# Cell 5 - EDA
print("Shape:", df.shape)
print("\nDescriptive stats:\n", df.describe())

# pairplot (saves figure)
plt.figure(figsize=(8,6))
sns.pairplot(df, hue="species", corner=True)
plt.suptitle("Pairplot of Iris features", y=1.02)
plt.savefig(os.path.join(FIG_DIR, "pairplot.png"), bbox_inches="tight")
plt.show()
```

```python
# Cell 6 - label encoding and train/test split
le = LabelEncoder()
y = le.fit_transform(df['species'])
X = df[["sepal_length","sepal_width","petal_length","petal_width"]].values
target_names = list(le.classes_)

print("Classes (label encoder):", target_names)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y, random_state=42)
print("Train/test sizes:", X_train.shape, X_test.shape)
```

```python
# Cell 7 - define candidate model pipelines
models = {
    "LogisticRegression": Pipeline([("scaler", StandardScaler()), ("clf", LogisticRegression(max_iter=300, random_state=42))]),
    "KNN": Pipeline([("scaler", StandardScaler()), ("clf", KNeighborsClassifier(n_neighbors=5))]),
    "DecisionTree": Pipeline([("clf", DecisionTreeClassifier(max_depth=5, random_state=42))]),
    "RandomForest": Pipeline([("clf", RandomForestClassifier(n_estimators=200, random_state=42))]),
    "SVM": Pipeline([("scaler", StandardScaler()), ("clf", SVC(probability=True, kernel='rbf', random_state=42))])
}
print("Models defined:", list(models.keys()))
```

```python
# Cell 8 - cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = {}
for name, pipe in models.items():
    scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy")
    cv_results[name] = scores
    print(f"{name:15s} : mean={scores.mean():.4f}, std={scores.std():.4f}")

# Boxplot of CV results
cv_df = pd.DataFrame(cv_results)
plt.figure(figsize=(8,5))
sns.boxplot(data=cv_df)
plt.title("Cross-validation accuracy distribution")
plt.ylabel("Accuracy")
plt.savefig(os.path.join(FIG_DIR, "cv_boxplot.png"), bbox_inches="tight")
plt.show()
```

```python
# Cell 9 - choose best model by mean CV and fit
best_name = max(cv_results.items(), key=lambda x: x[1].mean())[0]
best_pipe = models[best_name]
print("Best model by CV:", best_name)

best_pipe.fit(X_train, y_train)
y_pred = best_pipe.predict(X_test)
```

```python
# Cell 10 - evaluation metrics
acc = accuracy_score(y_test, y_pred)
prec_macro = precision_score(y_test, y_pred, average="macro")
rec_macro = recall_score(y_test, y_pred, average="macro")
f1_macro = f1_score(y_test, y_pred, average="macro")

print(f"Test Accuracy : {acc:.4f}")
print(f"Precision (macro) : {prec_macro:.4f}")
print(f"Recall (macro) : {rec_macro:.4f}")
print(f"F1 (macro) : {f1_macro:.4f}\n")
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=target_names))
```

```python
# Cell 11 - confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=target_names, yticklabels=target_names)
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.title(f"Confusion Matrix ({best_name})")
plt.savefig(os.path.join(FIG_DIR, "confusion_matrix.png"), bbox_inches="tight")
plt.show()
```

```python
# Cell 12 - ROC curves (one-vs-rest)
y_test_bin = label_binarize(y_test, classes=np.arange(len(target_names)))
n_classes = y_test_bin.shape[1]

# obtain probability scores
if hasattr(best_pipe, "predict_proba"):
    y_score = best_pipe.predict_proba(X_test)
else:
    try:
        y_score = best_pipe.decision_function(X_test)
    except Exception:
        y_score = None

if y_score is not None:
    plt.figure(figsize=(8,6))
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f"{target_names[i]} (AUC = {roc_auc:.2f})")
    plt.plot([0,1], [0,1], "k--", lw=1)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC curves (one-vs-rest)")
    plt.legend(loc="lower right")
    plt.savefig(os.path.join(FIG_DIR, "roc_multiclass.png"), bbox_inches="tight")
    plt.show()
else:
    print("No probability/decision scores available for ROC plotting.")
```

```python
# Cell 13 - feature importance or coefficients
features = ["sepal_length","sepal_width","petal_length","petal_width"]
plt.figure(figsize=(7,4))
if best_name in ("RandomForest", "DecisionTree"):
    clf = best_pipe.named_steps["clf"]
    importances = clf.feature_importances_
    sns.barplot(x=importances, y=features)
    plt.title("Feature importances")
    plt.xlabel("Importance")
    plt.savefig(os.path.join(FIG_DIR, "feature_importances.png"), bbox_inches="tight")
    plt.show()
elif best_name == "LogisticRegression":
    coefs = best_pipe.named_steps["clf"].coef_
    mean_abs = np.mean(np.abs(coefs), axis=0)
    sns.barplot(x=mean_abs, y=features)
    plt.title("Mean |coef| per feature (LogisticRegression)")
    plt.savefig(os.path.join(FIG_DIR, "feature_coefs.png"), bbox_inches="tight")
    plt.show()
else:
    print("Feature importance not available or not implemented for this model type.")
```

```python
# Cell 14 - learning curve
plt.figure(figsize=(8,6))
train_sizes, train_scores, val_scores = learning_curve(best_pipe, X, y, cv=cv, scoring="accuracy",
                                                      train_sizes=np.linspace(0.1, 1.0, 5), random_state=42)
train_mean = np.mean(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
plt.plot(train_sizes, train_mean, "o-", label="Training score")
plt.plot(train_sizes, val_mean, "o-", label="Cross-val score")
plt.xlabel("Training examples")
plt.ylabel("Score (accuracy)")
plt.title(f"Learning Curve ({best_name})")
plt.legend(loc="best")
plt.grid(True)
plt.savefig(os.path.join(FIG_DIR, "learning_curve.png"), bbox_inches="tight")
plt.show()
```

```python
# Cell 15 - save the best model + metadata
model_info = {
    "name": best_name,
    "model": best_pipe,
    "features": features,
    "target_names": target_names
}
MODEL_PATH = os.path.join(MODEL_DIR, f"best_iris_model_{best_name}.joblib")
joblib.dump(model_info, MODEL_PATH)
print("Saved model to:", MODEL_PATH)

# Example load (for future use)
# rec = joblib.load(MODEL_PATH)
# model = rec["model"]
# preds = model.predict(X_test)
```

```python
# Cell 16 - export dataset and write short report
df.to_csv(os.path.join(OUT_DIR, "iris_dataset_export.csv"), index=False)

report_text = f"""
Iris Classification Report
Best model: {best_name}
Test accuracy: {acc:.4f}
Precision (macro): {prec_macro:.4f}
Recall (macro): {rec_macro:.4f}
F1 (macro): {f1_macro:.4f}

Saved model: {MODEL_PATH}
Figures saved in: {FIG_DIR}
"""

with open(os.path.join(OUT_DIR, "report.txt"), "w") as f:
    f.write(report_text.strip())
print("Wrote report to:", os.path.join(OUT_DIR, "report.txt"))
print(report_text)
```
